```python
df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
```

```python
df.isnull().sum()
```

```python
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(" ","_")
df = df.rename(columns={"purchase_amount_(usd)" : "purchase_amount"})
```

```python
df.head()
```

```python
# create a column age_group
labels = ['Young Adult', 'Adult', 'Middle-aged', 'Senior']
df['age_group'] = pd.qcut(df['age'],q=4,labels=labels)
```

```python
df[['age','age_group']].head(10)
```

```python
import numpy as np
```

```python
# create column purchase_frequency_days
common_values = df['frequency_of_purchases'].value_counts().index.tolist()
print(common_values)
frequency_mapping = {}
for name in common_values:
    if name.lower() == "annually":
        frequency_mapping[name] = 365
    elif name.lower() == "every 3 months":
        frequency_mapping[name] = 90
    elif name.lower() == "quarterly":
        frequency_mapping[name] = 120
    elif name.lower() == "monthly":
        frequency_mapping[name] = 30
    elif name.lower() ==  'weekly':
        frequency_mapping[name] = 7
    elif name.lower() == "bi-weekly" or 'fortnightyly':
        frequency_mapping[name] = 14

frequency_mapping   
```

```python
df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
df[['purchase_frequency_days','frequency_of_purchases']].head(10)
```
